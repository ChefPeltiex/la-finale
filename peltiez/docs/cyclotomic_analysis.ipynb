# Analyse cyclotomique — racines primitives sur le cercle

Les racines **n**-ièmes de l’unité sont $\exp(2i\pi k/n)$ pour $k=0,\ldots,n-1$. Les **racines primitives** sont celles pour lesquelles $\gcd(k,n)=1$ (en particulier $k=0$ n’est primitif que si $n=1$, car $\gcd(0,n)=n$).

```python
import math
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np


def roots_plot(n, ax=None, save_svg=None):
    ks = list(range(n))
    prim_ks = [k for k in ks if math.gcd(k, n) == 1]
    non_prim_ks = [k for k in ks if math.gcd(k, n) != 1]
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    else:
        fig = ax.figure
    ax.set_aspect("equal")
    ax.axhline(0, color="k", lw=0.3)
    ax.axvline(0, color="k", lw=0.3)
    ax.add_patch(plt.Circle((0, 0), 1, fill=False, color="gray"))
    if non_prim_ks:
        th_np = 2 * np.pi * np.asarray(non_prim_ks) / n
        ax.scatter(
            np.cos(th_np),
            np.sin(th_np),
            c="#bbbbbb",
            s=40,
            zorder=2,
            label="non primitive",
        )
    th_p = 2 * np.pi * np.asarray(prim_ks) / n
    ax.scatter(
        np.cos(th_p),
        np.sin(th_p),
        c="#ff8c00",
        s=50,
        zorder=3,
        label="primitive (gcd(k,n)=1)",
    )
    ax.set_xlim(-1.15, 1.15)
    ax.set_ylim(-1.15, 1.15)
    ax.legend(loc="upper right", fontsize=8)
    ax.set_title(f"Racines n-ièmes, n={n}")
    if save_svg:
        fig.savefig(save_svg, format="svg", bbox_inches="tight")
    return fig, ax


n = 7
assets = Path("docs/assets")
assets.mkdir(parents=True, exist_ok=True)
save_svg = str(assets / f"roots_n{n}.svg")
fig, ax = roots_plot(n, save_svg=save_svg)
plt.show()
```
