```python
from sympy import Symbol, cancel, cyclotomic_poly, degree, divisors, expand, factor, mobius, prod


def cyclotomic_via_mobius(n, x):
    """Phi_n(x) = prod_{d|n} (x^d - 1)^(mu(n/d))."""
    if n < 1:
        raise ValueError("n >= 1")
    factors = []
    for d in divisors(n):
        exp = mobius(n // d)
        base = x**d - 1
        if exp == 0:
            continue
        if exp == 1:
            factors.append(base)
        elif exp == -1:
            factors.append(1 / base)
        else:
            factors.append(base**exp)
    return cancel(expand(prod(factors)))


X = Symbol("X")
for n in (7, 12, 105):
    p_mob = cyclotomic_via_mobius(n, X)
    p_sym = cyclotomic_poly(n, X, polys=False)
    assert expand(p_mob - p_sym) == 0
    print(f"Phi_{n}(X) = {p_sym}")
    print(f"  degre (phi({n})) = {degree(p_sym, X)}")
print("--- factorisation Phi_105 sur F_2 (style --factor) ---")
print(factor(cyclotomic_poly(105, X, polys=False), modulus=2))
```

```python
import math

import matplotlib.pyplot as plt


def plot_roots(ax, n, title):
    prim_x, prim_y = [], []
    other_x, other_y = [], []
    for k in range(n):
        ang = 2 * math.pi * k / n
        x, y = math.cos(ang), math.sin(ang)
        if math.gcd(k, n) == 1:
            prim_x.append(x)
            prim_y.append(y)
        else:
            other_x.append(x)
            other_y.append(y)
    ax.scatter(other_x, other_y, c="#7f8c8d", s=28, label="non primitive", zorder=2)
    ax.scatter(prim_x, prim_y, c="#e67e22", s=36, label="primitive", zorder=3)
    circ = plt.Circle((0, 0), 1, fill=False, color="#333")
    ax.add_patch(circ)
    ax.set_aspect("equal")
    ax.axhline(0, color="#ccc", lw=0.8)
    ax.axvline(0, color="#ccc", lw=0.8)
    ax.set_xlim(-1.15, 1.15)
    ax.set_ylim(-1.15, 1.15)
    ax.set_title(title)
    ax.legend(loc="upper right", fontsize=8)


fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, n in zip(axes, (7, 12, 105)):
    plot_roots(ax, n, f"n = {n}")
plt.tight_layout()
plt.show()
```
